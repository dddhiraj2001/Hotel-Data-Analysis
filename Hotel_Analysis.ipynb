```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
# Convert year, month, day to a single datetime column
df['date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + 
                            df['arrival_date_month'] + '-' + 
                            df['arrival_date_day_of_month'].astype(str))

# Sort by date
df = df.sort_values('date')

# Group by both date and hotel type, then calculate the mean of daily rates
daily_rates = df.groupby(['date', 'hotel'])['adr'].mean().unstack()

# Plotting
plt.figure(figsize=(20, 8))
plt.plot(daily_rates.index, daily_rates['City Hotel'], label='City Hotel')
plt.plot(daily_rates.index, daily_rates['Resort Hotel'], label='Resort Hotel')

plt.title('Average Daily Rate in City and Resort Hotels', fontsize=20)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Average Daily Rate (ADR)', fontsize=15)
plt.legend(title='Hotel Type')
plt.grid(True)
plt.show()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Extract month from the 'reservation_status_date'
df['month'] = df['reservation_status_date'].dt.month

# Plotting setup
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')

# Setting custom legend labels
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ['Not Canceled', 'Canceled'], title='Reservation Status', bbox_to_anchor=(1, 1))

plt.title('Reservation Status Per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per Month for Canceled Bookings', fontsize=30)

# Prepare the data
monthly_adr = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

# Creating a bar plot with a diverse color palette
# Here, 'husl' is a color palette that offers distinct colors. You can also use 'tab20' or define your own list of colors.
sns.barplot(x='month', y='adr', data=monthly_adr, palette='husl')

plt.xlabel('Month', fontsize=20)  # Adding an x-axis label
plt.ylabel('Total ADR', fontsize=20)  # Adding a y-axis label
plt.xticks(fontsize=15)  # Adjusting x-axis tick font size
plt.yticks(fontsize=15)  # Adjusting y-axis tick font size

plt.show()
```
